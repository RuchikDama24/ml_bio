```python
class DrugEmbeddingDataset(Dataset):
    """drug """

    def __init__(self, csv_file):
      self.data_frame = pd.read_csv(csv_file,header=None).to_numpy().astype(np.float32)


    def __len__(self):
      return len(self.data_frame)

    def __getitem__(self, idx):
        return self.data_frame[idx]
```

```python
def load_mapping(mapping_file):

	mapping = {}

	file_handle = open(mapping_file)

	for line in file_handle:
		line = line.rstrip().split()
		mapping[line[1]] = int(line[0])

	file_handle.close()

	return mapping
```

```python
tensor_transform = transforms.ToTensor()

dataset = DrugEmbeddingDataset("/content/drug2fingerprint.txt")

# DataLoader is used to load the dataset
# for training
loader = torch.utils.data.DataLoader(dataset = dataset,
                                     batch_size = 16,
                                     shuffle = True)
```

```python
# Creating a PyTorch class
# 2048 ==> 1024 ==> 512 ==> 128 ==> 64 ==> 32 ==> 10
class AE_Drugcell(torch.nn.Module):
	def __init__(self):
		super().__init__()

		# Building an linear encoder with Linear
		# layer followed by Relu activation function
		# 2048 ==> 10
		self.encoder = torch.nn.Sequential(
			torch.nn.Linear(2048, 1024),
			torch.nn.ReLU(),
			torch.nn.Linear(1024, 512),
      torch.nn.Dropout(p=0.2),
			torch.nn.ReLU(),
			torch.nn.Linear(512, 128),
      torch.nn.Dropout(p=0.2),
			torch.nn.ReLU(),
			torch.nn.Linear(128, 64),
			torch.nn.ReLU(),
			torch.nn.Linear(64, 32),
   		torch.nn.ReLU(),
      torch.nn.Linear(32,10)



		)

		# 10 ==> 2048
		self.decoder = torch.nn.Sequential(
      torch.nn.Linear(10,32),
      torch.nn.ReLU(),
			torch.nn.Linear(32, 64),
			torch.nn.ReLU(),
      torch.nn.Dropout(p=0.2),

			torch.nn.Linear(64, 128),
			torch.nn.ReLU(),
      torch.nn.Dropout(p=0.2),
			torch.nn.Linear(128, 512),
			torch.nn.ReLU(),
			torch.nn.Linear(512, 1024),
			torch.nn.ReLU(),
			torch.nn.Linear(1024, 2048),
			torch.nn.Sigmoid()
		)

	def forward(self, x):
		encoded = self.encoder(x)
		decoded = self.decoder(encoded)
		return decoded
```

```python
model = AE_Drugcell()
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
model = model.to(device)
# Validation using MSE Loss function
loss_function = torch.nn.BCELoss()

# Using an Adam Optimizer with lr = 0.1
optimizer = torch.optim.Adam(model.parameters(),
                             lr = 1e-1,
                             weight_decay = 1e-8)
```

```python
#boltzmann machine
```

```python
epochs = 100
outputs = []
losses = []
for epoch in range(epochs):
  for data in loader:
    data = data.to(device)
    reconstructed = model(data)

    # Calculating the loss function
    loss = loss_function(reconstructed, data)

    # The gradients are set to zero,
    # the the gradient is computed and stored.
    # .step() performs parameter update
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Storing the losses in a list for plotting
    losses.append(loss)
    #print(loss)
    outputs.append((epochs, data, reconstructed))

print(losses)
```

```python
print(losses)
for i in range(len(losses)):

  losses[i] = losses[i].cpu().detach().numpy()
plt.style.use('fivethirtyeight')
plt.xlabel('Iterations')
plt.ylabel('Loss')

# Plotting the last 100 values
plt.plot(losses[-100:])
plt.show()
```

```python
#https://debuggercafe.com/getting-started-with-variational-autoencoder-using-pytorch/
import torch
import torch.nn as nn
import torch.nn.functional as F
features = 16
# define a simple linear VAE
class LinearVAE(nn.Module):
    def __init__(self):
        super(LinearVAE, self).__init__()

        # encoder
        '''torch.nn.Linear(2048, 1024),
        torch.nn.ReLU(),
        torch.nn.Linear(1024, 512),
        torch.nn.Dropout(p=0.2),
        torch.nn.ReLU(),
        torch.nn.Linear(512, 128),
        torch.nn.Dropout(p=0.2),
        torch.nn.ReLU(),
        torch.nn.Linear(128, 64),
        torch.nn.ReLU(),
        torch.nn.Linear(64, 32),
        torch.nn.ReLU(),
        torch.nn.Linear(32,10)'''
        self.enc1 = nn.Linear(in_features=2048, out_features=1024)
        self.enc2 = nn.Linear(in_features=1024, out_features=512)
        self.enc3 = nn.Linear(in_features=512, out_features=128)
        self.enc4 = nn.Linear(in_features=128, out_features=10)

        # decoder
        self.dec1 = nn.Linear(in_features=10, out_features=128)
        self.dec2 = nn.Linear(in_features=128, out_features=512)
        self.dec3 = nn.Linear(in_features=512, out_features=1024)
        self.dec4 = nn.Linear(in_features=1024, out_features=2048)


    def reparameterize(self, mu, log_var):
        """
        :param mu: mean from the encoder's latent space
        :param log_var: log variance from the encoder's latent space
        """
        std = torch.exp(0.5*log_var) # standard deviation
        eps = torch.randn_like(std) # `randn_like` as we need the same size
        sample = mu + (eps * std) # sampling as if coming from the input space
        return sample

    def forward(self, x):
        # encoding
        x = F.relu(self.enc1(x))
        x = F.relu(self.enc2(x))
        x = F.relu(self.enc3(x))
        x = self.enc4(x)
        # get `mu` and `log_var`
        mu = x[:, 0, :] # the first feature values as mean
        log_var = x[:, 1, :] # the other feature values as variance
        # get the latent vector through reparameterization
        z = self.reparameterize(mu, log_var)

        # decoding
        x = F.relu(self.dec1(z))
        x = F.relu(self.dec2(x))
        x = F.relu(self.dec3(x))
        reconstruction = torch.sigmoid(self.dec4(x))
        return reconstruction, mu, log_var
```
